# Analysis overview

This project will present a comprehensive overview of Nordic startups that have reached unicorn status. A unicorn is a startup that has reached a valuation of atleast 1 billion US dollars pre IPO.




Methodology: All necessary data regarding these startups was sourced from the website 'Tracxn Technologies', a company that tracks startup data. Sources for each individual page will be provided in the bibliography section. All data is up to date as of June 16th 2025. The data that was relevant to the analysis was then manually transformed into the spreadsheet called "NordicUnicorns.csv". US dollars is the only currency used in this analysis


Limitations: As of July 2025, there are only 26 unicorns in Northern Europe. This relatively small sample size limits the depth of statistical analysis.

```python
Time_to_unicorn = [10.1, 8.34, 7.07, 8.18, 8.4]
Country_name = ['Sweden', 'Norway', 'Denmark', 'Finland', 'Iceland']


bars = plt.bar(Country_name, Time_to_unicorn, color="#27AE60", edgecolor='black')
plt.title("Average time to unicorn by country")
plt.ylabel("Time to unicorn (years)")
plt.xlabel("Country")
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.15, f'{height:.2f}',
             ha='center', va='bottom', fontsize=10)
```

 Sweden has the highest average time to unicorn status at 10.1 years.Denmark leads in speed, with unicorns reaching status in just 7.1 years on average. The remaining countries (Norway, Finland, Iceland) hover at around 8.2-8.4 years. This could indicate that regional factors like funding access, market size, or regulation influence the average time more than national policies do with the exception of Sweden and Denmark, which are outliers.

# Distribution by industries

For a better analysis, a more broad definition of each industry was created. So that for example: cloud analytics and cloud infrastructure fall under the same category of enterprise SaaS.

```python

industry_map = {
    "Fintech": "Fintech",
    "Neobank": "Fintech",
    "HealthTech": "Health and Biotech",
    "HealtTech": "Health and Biotech",
    "Biotech": "Health and Biotech",
    "Consumer Electronics": "Consumer Tech",
    "Gaming": "Consumer Tech",
    "Streaming": "Consumer Tech",
    "B2B SaaS": "Enterprise SaaS",
    "Retail SaaS": "Enterprise SaaS",
    "SaaS/ E-commerce infrastructure": "Enterprise SaaS",
    "Cloud Infrastructure": "Enterprise SaaS",
    "Crypto Analytics": "Enterprise SaaS",
    "GreenTech": "Green and Industrial",
    "Industrial AI": "Green and Industrial",
    "Logistics": "Green and Industrial",
    "E-commerce logistics": "Green and Industrial",
    "FoodTech": "Other",
    "Transport": "Other",
    "Enterprise SaaS": "Enterprise SaaS"
}

df["Broad Industry"] = df["Industry"].map(industry_map)
df.loc[:,'Broad Industry']
```

```python
category_counts = df['Broad Industry'].value_counts()
plt.pie(category_counts, labels=category_counts.index, autopct="%.1f%%")
plt.title('Nordic Unicorns distribution by industries')
plt.show()
```

The unicorn landscape in the Nordic region is very diverse with 5 broad industries. Enterprise SaaS dominates with a total of 23.1% (6 out of 26)  of all nordic unicorns falling under this category. Notably, Health and Biotech also share decently large percentage of the unicorns (15.4%).This is likely caused by the fact that Nordic countries have a very strong healthcare system with universal healthcare being completely free in all.

# Number of unicorns by country

```python
#The population of the countries varies signifficantly, so to compare the the number of unicorns per country, we will
```

The population of the countries varies signifficantly, so to compare the the number of unicorns per country, we will divide it by the number of people per country as well as the number of startups.
There are a total of 26.1K startups in Sweden
13.4K startups in Norway
13.2K startups in finland

```python
df2 = pd.DataFrame([["Sweden",12, 2.61, 10.6],["Norway",5,1.34,5.51],["Denmark",3, 1.86,5.97],["Finland",5, 1.32,5.63],["Iceland",1, 0.743,0.36]],
                   columns=["Country","Number of Unicorns", "Number of startups(thousands)", "Number of people(millions)"])
df2.loc[:, "Unicorns per M people"] = df2.loc[:, "Number of Unicorns"] / df2.loc[:, "Number of people(millions)"]
df2.loc[:, "Unicorns per ten thousand startups"] = df2.loc[:, "Number of Unicorns"] / df2.loc[:, "Number of startups(thousands)"]
df2
```

```python
plt.bar(df2["Country"], df2["unicorns per M people"], color="skyblue")
plt.xlabel("Country")
plt.ylabel("Number of unicorns")
plt.title("Unicorns per M people")
```

Iceland stands out with the highest unicorn density with 2.78 unicorns per 1 million people. It is however, important to note that the population of Iceland is 364 thousand and only 1 unicorn has come out of Iceland so the sample size is too small to draw any meaningful conclusions.

```python
plt.bar(df2["Country"], df2["unicorns per ten thousand startups"], color="green")
plt.xlabel("Country")
plt.ylabel("Number of unicorns")
plt.title("Unicorns per ten thousand startups")
```

Sweden leads in startup quality, with 4.6 unicorns per 10k startups. indicating better conversion rates from startups to unicorns. Norway and Finland show a roughly equal performance with 3.7 unicorns per 10k startups. Denmark falls behind with only 1.6 unicorns per 10k startups. Iceland is last with 1.3 unicorns per 10k startups.
